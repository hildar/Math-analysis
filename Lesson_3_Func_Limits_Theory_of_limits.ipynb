### Задача 4

Найти предел:
$\lim \limits_{x \to 0} \dfrac{3x^3 - 2x^2}{4x^2} $

#### Решение:

$\lim \limits_{x \to 0} \dfrac{3x^3 - 2x^2}{4x^2} = \left( \dfrac{0}{0} \right) = \dfrac{x^2 (3x - 2)}{4x^2} = \dfrac{3x - 2}{4} = \dfrac{0 - 2}{4} = -\dfrac{1}{2} $

### \*Задача 5

Найти предел:
$\lim \limits_{x \to 0} \dfrac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1} $

#### Решение:

$\lim \limits_{x \to 0} \dfrac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1} = \left( \dfrac{0}{0} \right) = \dfrac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1} \cdot \dfrac{(\sqrt[3]{1 + x})^2 + (\sqrt[3]{1 + x})\cdot 1 + 1^2}{(\sqrt[3]{1 + x})^2 + (\sqrt[3]{1 + x})\cdot 1 + 1^2} = \dfrac{(\sqrt{1 + x} - 1) \cdot ((\sqrt[3]{1 + x})^2 + (\sqrt[3]{1 + x})\cdot 1 + 1^2)}{(\sqrt[3]{1 + x})^3 - 1^3} = $

$= \dfrac{(\sqrt{1 + x} - 1) \cdot ((\sqrt[3]{1 + x})^2 + (\sqrt[3]{1 + x})\cdot 1 + 1)}{x} \cdot \dfrac{\sqrt{1 + x} + 1}{\sqrt{1 + x} + 1} = \dfrac{((\sqrt{1 + x})^2 - 1) \cdot ((\sqrt[3]{1 + x})^2 + (\sqrt[3]{1 + x})\cdot 1 + 1)}{x \cdot \sqrt{1 + x} + 1} = $

$= \dfrac{x \cdot ((\sqrt[3]{1 + x})^2 + \sqrt[3]{1 + x} + 1)}{x \cdot \sqrt{1 + x} + 1} = \dfrac{(\sqrt[3]{1 + x})^2 + \sqrt[3]{1 + x} + 1}{\sqrt{1 + x} + 1} = \dfrac{(\sqrt[3]{1 + 0})^2 + \sqrt[3]{1 + 0} + 1}{\sqrt{1 + 0} + 1} = \dfrac{1 + 1 + 1}{1 + 1} = \dfrac{3}{2} $

### \*Задача 6

Найти предел:
$\lim \limits_{x \to \infty} \left( \dfrac{x + 3}{x} \right)^{4x + 1} $

#### Решение:

Для решения неопределенности приведем функцию ко второму замечательному пределу:

$\lim \limits_{x \to \infty} \left( \dfrac{x + 3}{x} \right)^{4x + 1} = \left( \dfrac{\infty}{\infty} \right)^{\infty} = \left( \left( 1 + \dfrac{1}{\frac{x}{3}} \right)^{\frac{x}{3}} \right)^{\frac{3}{x}\cdot(4x + 1)} = e^{\lim \limits_{x \to \infty} \frac{3}{x}\cdot(4x + 1) } = e^{\lim \limits_{x \to \infty} \frac{12x + 3}{x}}  $

Теперь осталось решить предел $\lim \limits_{x \to \infty} \frac{12x + 3}{x}$ т.к. он тоже имеет неопределенность вида $\left( \dfrac{ \infty}{\infty} \right) $. Для этого разделим каждый член числителя и знаменателя на $x$:

$\lim \limits_{x \to \infty} \dfrac{12x + 3}{x} = \left( \dfrac{ \infty}{\infty} \right) = \dfrac{12 + \frac{3}{x}}{1} = \dfrac{12 + \frac{3}{\infty}}{1} = \dfrac{12 + 0}{1} = 12$

Соответственно ответ:

$\lim \limits_{x \to \infty} \left( \dfrac{x + 3}{x} \right)^{4x + 1} = e^{12}$

## Теоремы о пределах
### Задача 7.1

Найти предел:
$\lim \limits_{x \to 0} \dfrac{sin(2x)}{4x} $

#### Решение:

Приведем функцию к первому замечательному пределу и решим:

$\lim \limits_{x \to 0} \dfrac{sin(2x)}{4x} = \left( \dfrac{0}{0} \right) = \dfrac{sin(2x)}{2\cdot 2x} = \dfrac{1}{2}\cdot \dfrac{sin(2x)}{2x} = \dfrac{1}{2}$

### Задача 7.2

Найти предел:
$\lim \limits_{x \to 0} \dfrac{x}{sin(x)} $

#### Решение:

Данный лимит есть ничто иное как первый замечательный предел:

$\lim \limits_{x \to 0} \dfrac{x}{sin(x)} = \left( \dfrac{0}{0} \right) = 1$

### Задача 7.3

Найти предел:
$\lim \limits_{x \to 0} \dfrac{x}{arcsin(x)} $

#### Решение:

При $x \rightarrow 0$, $arcsin(x) \approx x$

$\lim \limits_{x \to 0} \dfrac{x}{arcsin(x)} = \dfrac{x}{x} = 1$

### Задача 7.4

Найти предел:
$\lim \limits_{x \to \infty} \left( \dfrac{4x + 3}{4x - 3} \right)^{6x} $

#### Решение:

Для решения неопределенности приведем функцию ко второму замечательному пределу:

$\lim \limits_{x \to \infty} \left( \dfrac{4x + 3}{4x - 3} \right)^{6x} = \left( \dfrac{\infty}{\infty} \right)^{\infty} = \left( \dfrac{4x - 3 + 6}{4x - 3} \right)^{6x} = \left( 1 + \dfrac{1}{\frac{4x - 3}{6}} \right)^{6x} = \left( \left( 1 + \dfrac{1}{\frac{4x - 3}{6}} \right)^{\frac{4x - 3}{6}} \right)^{\frac{6}{4x - 6}\cdot 6x} = e^{\lim \limits_{x \to \infty} \frac{36x}{4x - 3}}$

Теперь осталось решить предел $\lim \limits_{x \to \infty} \frac{36x}{4x - 3}$ т.к. он тоже имеет неопределенность вида $\left( \dfrac{ \infty}{\infty} \right) $. Для этого разделим каждый член числителя и знаменателя на $x$:

$\lim \limits_{x \to \infty} \dfrac{36x}{4x - 3} = \left( \dfrac{ \infty}{\infty} \right) = \lim \limits_{x \to \infty} \dfrac{\frac{36x}{x}}{\frac{4x}{x} - \frac{3}{x}} = \dfrac{36}{4 - \frac{3}{x}} = \dfrac{36}{4 - \frac{3}{\infty}} = \dfrac{36}{4 - 0} = 9$

Соответственно ответ:

$\lim \limits_{x \to \infty} \left( \dfrac{4x + 3}{4x - 3} \right)^{6x}  = e^{9}$

### \*Задача 7.5

Найти предел:
$\lim \limits_{x \to \infty} \dfrac{sin(x) + ln(x)}{x}$

#### Решение:

$\lim \limits_{x \to \infty} \dfrac{sin(x) + ln(x)}{x} = \left( \dfrac{sin(x)}{x} + \dfrac{ln(x)}{x} \right) = \left( 1 + \dfrac{ln(x)}{x} \right) = \left( 1 + \dfrac{(ln(x))'}{(x)'} \right) = \left( 1 + \dfrac{\frac{1}{x}}{1} \right) = \left( 1 + \dfrac{1}{\infty} \right) = 1$
